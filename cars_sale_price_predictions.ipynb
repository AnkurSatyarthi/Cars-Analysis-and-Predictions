# Cars Sale Price Predictions

```python
import pandas as pd
cars = pd.read_excel('data/cars_dataset.xlsx')
%matplotlib inline
# Relationship b/n years and selling prices
cars.plot(kind='scatter',x='year',y='selling_price')
```

```python
# Relationship b/n km driven and selling prices
cars.plot(kind='scatter',x='km_driven',y='selling_price')
```

```python
response = 'selling_price'
y = cars[[response]]
y
```

```python
predictors = list(cars.columns)
unnamed = 'Unnamed: 0'
predictors.remove(response)
predictors.remove(unnamed)
x=cars[predictors]
x
```

```python
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
x_train,x_test,y_train,y_test = train_test_split(x,y,random_state=1234)
selling_price_prediction_model = LinearRegression().fit(x_train,y_train)
selling_price_prediction_model.intercept_
```

```python
selling_price_prediction_model.coef_
```

```python
# Evaluation
selling_price_prediction_model.score(x_test, y_test)
```

```python
from sklearn.metrics import mean_absolute_error
y_pred = selling_price_prediction_model.predict(x_test)
mean_absolute_error(y_test,y_pred)
```
