# Car Sales Analysis

```python
import pandas as pd
cars = pd.read_csv('data/CAR DETAILS FROM CAR DEKHO.csv')
cars.head()
```

```python
%matplotlib inline
```

## 1. Relationships

### Objective
**1.**  **Relationship between Car's year and their selling prices**

```python
cars.plot(kind='scatter',x='year',y='selling_price')
```

**2.**  **Relationship between Car's km_driven and their selling prices**

```python
cars.plot(kind='scatter',x='km_driven',y='selling_price')
```

**3.**  **Relationship between Car's owner and their selling prices**

```python
cars.plot(kind='scatter',x='owner',y='selling_price',figsize=(9,8))
```

```python
cars.plot(kind='scatter',x='fuel',y='selling_price')
```

```python
cars.plot(kind='scatter',x='seller_type',y='selling_price')
```

```python
cars.groupby('seller_type')['owner'].value_counts().unstack().plot(kind='bar',stacked=True,figsize=(10,6))
```

```python
cars.groupby('transmission')['owner'].value_counts().unstack().plot(kind='bar',stacked=True,figsize=(10,6))
```

```python
cars.groupby('fuel')['owner'].value_counts().unstack().plot(kind='bar',stacked=True,figsize=(10,6))
```
